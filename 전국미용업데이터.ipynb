```python
def optional_index(_list, index):
    try:
        return _list[index]
    except IndexError:
        return None
```

```python
도로명 = pd.DataFrame({i:data["도로명주소_split"].apply(lambda x: optional_index(x, i)) for i in range(18)})
```

```python
def icontains(keyword):
    return lambda a : keyword in a
def series_contains(series, keyword):
    return series.astype("str").apply(icontains(keyword))
def iter_series_contains(iter, keyword, iter_count):
    return [series_contains(iter[i], keyword) for i in range(iter_count)]
def iterOR(iter):
    default = iter[0]
    for it in iter[1:]:
        default = default | it
    return default

def iter_series_contains_or(iter, keyword, iter_count):
    return iterOR(iter_series_contains(iter, keyword, iter_count))

def iterOR_func(iter, func):
    default = iter[0].astype("str").apply(func)
    for it in iter[1:]:
        default = default | it.astype("str").apply(func)
    return default
```

```python
분당미용실 = 도로명.loc[iterOR_func([도로명[i] for i in range(3)], icontains("분당"))]
별내미용실 = 도로명.loc[iterOR_func([도로명[i] for i in range(3)], icontains("별내"))]
중계미용실 = 도로명.loc[iterOR_func([도로명[i] for i in range(3)], icontains("중계"))]
강남미용실 = 도로명.loc[iterOR_func([도로명[i] for i in range(3)], icontains("강남")) & 도로명[0].astype("str").apply(icontains("서울특별시"))]
조치원미용실 = 도로명.loc[iter_series_contains_or(도로명, "조치원", 3) & iter_series_contains_or(도로명, "세종", 1)]
```
